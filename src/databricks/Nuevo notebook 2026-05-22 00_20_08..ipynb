```python
# Leer el CSV (sin SAS Token, ya no lo necesitas)
df_gps = (spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv("abfss://raw-zone@stdatacolake.dfs.core.windows.net/GPS/Datos_vehiculos.csv")
)

display(df_gps)
```

```python
# Guardar como Delta en Bronze
df_gps.write \
    .format("delta") \
    .mode("overwrite") \
    .save("abfss://raw-zone@stdatacolake.dfs.core.windows.net/bronze/gps")

print("✅ Guardado en Delta!")

# Verificar que se creó
dbutils.fs.ls("abfss://raw-zone@stdatacolake.dfs.core.windows.net/bronze/gps")
```
